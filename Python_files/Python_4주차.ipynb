```python
#직접해보기
#10개의 수를 입력받아 리스트를 생성하고 뒤에서부터 거꾸로 출력하세요.
#정답

madelist = []

for i in range(0, 10) :
  num = int(input(str(i) + "번째 숫자를 입력하세요."))
  print(num)
  madelist.append(num)

madelist.reverse()

print(madelist)
```
