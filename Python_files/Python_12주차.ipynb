반복문: while

```python
i = 0

while (i < 10):
  print("안녕하세요.")
  i += 1
```

```python
sum = 0
n1, n2 = 0, 0

while True:
  n1 = int(input())
  n2 = int(input())
  sum = n1 + n2
  print("두 수의 합은", sum, "입니다.")
```

```python
i = 0

while (i < 10):
  print("안녕하세요.")
  i += 1
  if i == 5:
    break;
```

```python
for i in range(10):
  if i == 5:
    break;
  print("안녕하세요.")
```

```python
sum = 0
n1, n2 = 0, 0

while True:
  print("0을 입력하면 프로그램을 마칩니다.")
  n1 = int(input())
  if n1 == 0:
    break
  n2 = int(input())
  sum = n1 + n2
  print("두 수의 합은", sum, "입니다.")
```

```python
i = 0

while(i < 10):
  i += 1
  if i == 5:
    continue
  print(i, "번째 반복입니다.")
```

```python
for i in range(10):
  if i == 5:
    continue
  print(i, "번째 반복입니다.")
```

```python
#직접해보기

for i in range(2, 10, 1):
  for k in range(2, 10, 1):
    print(i, "*", k, "=", i * k)
```

```python
#직접해보기

import random

i = 1

while(i < 6):
  print(i, "번째 게임")
  print("컴퓨터가 생각하는 숫자는?(1~5)")

  user = int(input())
  com = random.randint(1, 5)

  if user == com:
    print("맞추었네요! 축하드려요.")
    break
  else:
    print("틀렸습니다. 컴퓨터가 생각하는 숫자는", com, "입니다.")
  i += 1
```
