반복문

```python
for i in range(0, 10, 1):
  print("안녕하세요")
```

```python
for i in range(10):
  print("안녕하세요.")
```

```python
for i in range(10):
  print("안녕하세요.")
  print("만나서 반갑습니다.")
```

```python
for i in range (0,10, 1):
  print(i, "안녕하세요.")
```

```python
for i in range(0, 10, 1):
  print(i, "번 손님입니다.")
```

```python
sum = 0

for i in range (1, 11, 1):
  sum += i
print("1부터 10까지의 합은", sum, "입니다.") #이 문장을 반복하고 싶지 않기 때문에 블록 밖에 위치하도록 해야 함.
```

```python
sum = 0

for i in range(1, 500, 1):
  if i % 2 == 0:
    sum += i

print("1부터 500까지의 수 중 짝수의 합은", sum, "입니다.")
```

```python
sum = 0

for i in range(2, 500, 2):
  sum += i

print("1부터 500까지의 수 중 짝수의 합은", sum, "입니다.")
```

```python
#직접해보기

for i in range(2, 6, 1):
  print(i, end = '')
```

```python
#직접해보기
sum = 0

for i in range(-2023, 0, 1):
  sum += i

print("-1부터 -2023까지의 합은", sum, "입니다.")
```

```python
#직접해보기

for i in range(1, 2007, 1):
  if i % 5 == 0 :
    print(i)
```
