#대입연산자

```python
n = 100
print(n)
n = 100 + 200
print(n)
n = "Hi" + "Hello"
print(n)
```

```python
n1, n2 = 100, 200
print(n1)
print(n2)
n1, n2 = 100 + 200, 100 * 200
print(n1)
print(n2)
n1, n2 = "Hi" + "Hello", "Hello"+"Hi"
print(n1)
print(n2)
```

```python
n1, n2, n3 = 2, 3, 6
print(n1, n2, n3)
```

```python
#직접해보기
#파운드를 kg으로 변환해주는 프로그램을 만드세요.
#조건: 1파운드(lb) = 0.453592킬로그램(kg), 1kg = 2.204623lb

lb = float(input())
print("파운드를 입력하세요 : ", lb)
lb *= 0.453592

print(float(input()), "파운드(lb)는", lb, "킬로그램(kg) 입니다.")
```

```python
#정답
pound = int(input("파운드(lb)를 입력하세요 : "))
kg = pound * 0.453592
print(pound, "파운드(lb)는", kg, "킬로그램(kg)입니다.")
```

```python
#직접해보기
#연산자 우선순위에 주의하여 이수학점에 따른 평균학점을 구하는 프로그램을 만드시오.
# 파이썬(3) B(3.5), C언어(2) A(4.0), 모바일(1) A+(4.5)

python = 3.5 * 3
C = 4.0 * 2
mobile = 4.5

avg = (float(python) + float(C) + float(mobile)) / 6
print("평균 학점: " , avg)
```

```python
python = 3
c = 2
mobile = 1
a, aa, aaa = 3.5, 4.0, 4.5

avg = ((python*a) + (c*aa) + (mobile*aaa)) / (python + c + mobile)
print("평균 학점: " , avg)
```
